```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import csv
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

SCAP CNBC

```python
def scrape_cnbc(hal):
    global hades
    a = 1
    with open('cnbc.csv', 'a', encoding='utf-8', newline='') as file:
        wr = csv.writer(file, delimiter=',')
        wr.writerow(['title', 'text', 'subject', 'date'])  # Menambahkan baris header ke file CSV
        for page in range(1, hal):
            url_cbnc = f'https://www.cnbcindonesia.com/search?query=kejaksaan%20agung&p={page}&kanal=&tipe=artikel&date= '  
            ge = req.get(url_cbnc, hades).text
            sop = bs(ge, 'lxml')
            ul = sop.find ('ul', class_ = 'list media_rows middle thumb terbaru gtm_indeks_feed')
            li = ul.find_all('li')
            for x in li:
                link = x.find('article').find('a')['href']
                ge_ = req.get(link, hades).text
                sop_ = bs(ge_, 'lxml')
                art = sop_.find ('div', class_ = 'lm_content mt10').find('article')
                headline = art.find ('h1').text
                date = art.find('div', class_='date').text
                date = datetime.strptime (date, '%d %B %Y %H:%M')              
                content = sop_.find_all('div', class_='detail_text')
                for x in content:
                    x = x.find_all('p')
                    y = [y.text for y in x]
                    content_ = ''.join(y).replace('/n', '').replace('ADVERTISEMENT', '').replace('SCROLL TO RESUME CONTENT', '').replace('SCROLL TO CONTINUE WITH CONTENT','')
                    # Menghilangkan tanda koma 
                    headline = headline.replace(',', '')
                    content_ = content_.replace(',', '')
                    
                    
                    print(f'done[{a}] > {headline[0:10]}')
                    a += 1
                    wr.writerow([headline, content_,'News', date])
```

```python
scrape_cnbc(3)
```

SCRAP DETIK

```python
def scrape_detik(hal):
    global hades
    a = 1
    with open('detik.csv', 'a', encoding='utf-8', newline='') as file:
        wr = csv.writer(file, delimiter=',')
        wr.writerow(['title', 'text', 'subject', 'date'])  # Menambahkan baris header ke file CSV
        for page in range(1, hal):
            url_detik = f'https://www.detik.com/search/searchnews?query=Kejaksaan+Agung&sortby=time&page={page}'
            ge = req.get(url_detik, hades).text
            sop = bs(ge, 'lxml')
            li = sop.find('div', class_='list media_rows list-berita')
            lin = li.find_all('article')
            for x in lin:
                link = x.find('a')['href']
                date = x.find('a').find('span', class_='date').text.replace('WIB', '').replace('detikNews', '').split(',')[1]
                
                # Konversi format tanggal "mm/dd/yyyy"
                date_obj = datetime.strptime(date.strip(), '%d %b %Y %H:%M')
                date_formatted = datetime.strftime(date_obj, '%m/%d/%Y')

                headline = x.find('a').find('h2').text
                ge_ = req.get(link, hades).text
                sop_ = bs(ge_, 'lxml')
                content = sop_.find_all('div', class_='detail__body-text itp_bodycontent')
                for x in content:
                    x = x.find_all('p')
                    y = [y.text for y in x]
                    content_ = ''.join(y).replace('/n', '').replace('ADVERTISEMENT', '').replace('SCROLL TO RESUME CONTENT', '').replace('SCROLL TO CONTINUE WITH CONTENT','')

                    # Menghilangkan tanda koma 
                    headline = headline.replace(',', '')
                    content_ = content_.replace(',', '')
                    date = date.replace(',', '')

                    print(f'done[{a}] > {headline[0:10]}')
                    a += 1
                    wr.writerow([headline, content_,'News', date_formatted])
```

```python
scrape_detik(3)
```
